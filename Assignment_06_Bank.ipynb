```python
cust_info={}
bank_name='RBL'
width=60

def creat_new_account():
    global cust_info
    name=input('Enter your name : ')
    Accno=input('Enter New Account Number : ')
    if Accno in cust_info:
        print('ERROR : Account no already exists ')
    else:
        password=input('Enter your password : ')
        cust_info[Accno]={'name':name,'bal':0,'pass':password}
        print(f'New Account number {Accno} opened Successfully ')

def get_balance(Accno):
    return cust_info[Accno]['bal']

def deposit(Accno,amount):
    cust_info[Accno]['bal'] = int(amount) + cust_info[Accno]['bal']

    
def withdraw(Accno,amount):
    if cust_info[Accno]['bal'] >=amount:
        cust_info[Accno]['bal'] = cust_info[Accno]['bal']-amount
    else:
        print('ERROR : Insufficient funds')

def header():
    hash = '-' * width
    print(f'{hash}\n{bank_name.center(width)}\n{hash}')
    
def main_screen():
    header()
    valid_choices = ['1','2']
    print('1)Login\n2)New User')
    choice =input ('Enter valid choice : ')
    if choice not in valid_choices:
        print('ERROR : Invalid choice')
        return
    return choice

def get_user_credentials():
    Accno=input('Enter Account Number : ')
    password=input('Enter password : ')
    return Accno,password

def verify_credentials(Accno,password):
    '''
    This function verifies credentials
    param Account number: account number
    param password :account password
    return : Bool
    '''
    if Accno in cust_info and cust_info[Accno]['pass']==password:
        return True
    return False

def get_logged_in_user_choice():
    header()
    valid_choices=['1','2','3']
    print('1)Get Balance\n2)Deposite\n3)Withdraw')
    choice=input('Enter choice : ')
    if choice not in valid_choices:
        print('Error : Invalid choice')
        return
    return choice

def handle_logged_in_user(Accno):
    while True:
        choice=get_logged_in_user_choice()
        if choice=='1':
            print(f'{cust_info[Accno]["name"]} your account balance is {get_balance(Accno)}')
        elif choice=='2' or choice=='3':
            amt=input('Enter amount : ')
            if amt.isdigit():
                if choice=='2':
                    deposit(Accno,int(amt))
                else:
                    withdraw(Accno,int(amt))
            else:
                print('ERROR : Invalid amount')
        else:
            break
def server():
    while True:
        choice=main_screen()
        if choice=='1':
            Accno,password=get_user_credentials()
            if verify_credentials(Accno,password):
                handle_logged_in_user(Accno)
            else:
                print('Error : Invalid login credentials')
        elif choice=='2':
            creat_new_account()
        else:
            break
```
