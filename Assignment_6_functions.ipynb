```python
cust_info={'123':{'name':'Ethans',
                 'city':'Pune',
                 'balance':100,
                 'password':123
                 }
          }
def check_account_exists(account_no):
    if account_no in cust_info:
        return True
    else:
        return False
def create_account(name,city,account_no,password):
    global cust_info
    if not check_account_exists(account_no):
        cust_info[account_no]={'name':name,
                              'city':city,
                              'balance':0,
                              'password':password
                              }
        print(f'{name} your account created successfully')
    else:
        print('ERROR:Account Number Alredy Exists')
        
def vrify_password(account_no,password):
    acc_exists=check_account_exists(account_no)
    if acc_exists and cust_info[account_no]['password']==password:
        return True
    return False

def get_balance(account_no,password):
    if vrify_password(account_no,password):
        name=cust_info[account_no]['name']
        bal=cust_info[account_no]['balance']
        print(f'{name} your balance is {bal}')
    else:
         print('ERROR:Account Number Alredy Exists')
            
def deposit(account_no,amount):
    global cust_info
    if check_account_exist(account_no):
        name=cust_info[account_no]['name']
        bal=cust_info[account_no]['balance']
        cust_info[account_no]['balance']=bal+amount
        print(f"{name} your updated balance is {cust_info[account_no]['balance']}")
    else:
         print('ERROR:Account Number Alredy Exists')
            
def withdraw(account_no,amount):
    global cust_info
    if check_account_exist(account_no):
        name=cust_info[account_no]['name']
        bal=cust_info[account_no]['balance']
        if bal>=amount:
            cust_info[account_no]['balance']=bal-amount
            print(f"{name} your updated balance is {cust_info[account_no]['balance']}")
        else:
            print(f'ERROR :{name} you have insufficient balance')
    else:
         print('ERROR:Account Number Alredy Exists')
    
        
```
